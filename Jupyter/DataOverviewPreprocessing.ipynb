```python
import pandas as pd
```

```python
# 1. Load the dataset
file_path = "EEG.machinelearing_data_BRMH.csv"
df = pd.read_csv(file_path)
```

```python
# 2. Count the number of rows and columns
# The .shape attribute returns a tuple: (number of rows, number of columns)
num_rows, num_cols = df.shape
print(f"Total number of rows (samples): {num_rows}")
print(f"Total number of columns (attributes): {num_cols}")
print("-" * 30)
```

```python
# 3. Print all the column names
print("List of all columns in the dataset:")
# We convert it to a list so it prints nicely, 
# which is helpful when you have ~1,000 columns!
column_names = list(df.columns)
print(column_names)
print("-" * 30)
```

```python
# .info() tells you the data types (numbers, text) and if there are missing values
print("\nDataset Info:")
df.info()
```

```python
print("=== DISORDER BREAKDOWN ===")
# 2. Get unique main disorders
main_disorders = df['main.disorder'].dropna().unique()
print(f"Total Unique Main Disorders: {len(main_disorders)}")

# 3. Group specific disorders under their main disorder
for main_d in main_disorders:
    # Filter the dataframe for this main disorder, then find unique specific disorders
    sub_disorders = df[df['main.disorder'] == main_d]['specific.disorder'].dropna().unique()
    print(f"\nMain Disorder: {main_d}")
    print(f"  -> Specific Disorders: {', '.join(sub_disorders)}")

print("\n" + "="*30 + "\n")
```

```python
print("=== EEG COLUMN PARSING ===")
# 4. Extract all possible subparts from the EEG columns
# The EEG columns start after 'specific.disorder'. 
# Let's find the index of 'specific.disorder' and take everything after it.
start_idx = list(df.columns).index('specific.disorder') + 1
eeg_columns = df.columns[start_idx:]

# We use Sets because they automatically store only UNIQUE values
measurement_types = set()
frequencies = set()
channels = set()

for col in eeg_columns:
    parts = col.split('.')
    
    # 1st part is always Measurement Type (AB or COH)
    measurement_types.add(parts[0])
    
    # 3rd part is always Frequency (delta, gamma, etc.)
    # (The 2nd part is just a grouping letter like 'A' or 'F')
    if len(parts) >= 3:
        frequencies.add(parts[2])
    
    # Extracting channels (electrodes)
    if parts[0] == 'AB' and len(parts) >= 5:
        # AB format: AB.A.delta.a.FP1 -> Channel is the 5th part (index 4)
        channels.add(parts[4])
        
    elif parts[0] == 'COH' and len(parts) >= 7:
        # COH format: COH.A.delta.g.F8.h.T3 -> Channels are 5th (index 4) and 7th (index 6)
        channels.add(parts[4])
        channels.add(parts[6])

# Print the extracted unique components
print(f"Unique Measurement Types ({len(measurement_types)}):")
print(sorted(list(measurement_types)))

print(f"\nUnique Frequency Bands ({len(frequencies)}):")
print(sorted(list(frequencies)))

print(f"\nUnique Electrode Channels ({len(channels)}):")
print(sorted(list(channels)))
```

```python


# 2. Clean up (Drop any accidental 'Unnamed' columns)
unnamed_cols = [col for col in df.columns if 'Unnamed' in col]
if unnamed_cols:
    df = df.drop(columns=unnamed_cols)
    print(f"Cleaned up {len(unnamed_cols)} empty/unnamed columns.\n")

# 3. Check Dataset Balance (How many patients per disorder?)
print("=== PATIENT COUNTS BY MAIN DISORDER ===")
disorder_counts = df['main.disorder'].value_counts()
print(disorder_counts)
print("-" * 40)

# 4. Demographics Check (Age and Sex)
print("=== AVERAGE AGE BY DISORDER ===")
# Group by the disorder and calculate the average age, rounded to 1 decimal
age_by_disorder = df.groupby('main.disorder')['age'].mean().round(1)
print(age_by_disorder)
print("-" * 40)

print("=== SEX DISTRIBUTION ===")
# Assuming sex is coded (e.g., M/F or 0/1)
sex_counts = df['sex'].value_counts()
print(sex_counts)
```
