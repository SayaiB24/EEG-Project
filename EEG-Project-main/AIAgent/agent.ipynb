```python
def get_level(score):
    if score <= 25:
        return "Minimal / Normal"
    elif score <= 50:
        return "Moderate"
    elif score <= 75:
        return "Severe"
    else:
        return "Extremely Severe"
```

```python
depression = 35
anxiety = 60
stress = 70
attention = 40
eeg_attention = 42
confidence = 55
```

```python
depression_label = f"{get_level(depression)} ({depression}%)"
anxiety_label = f"{get_level(anxiety)} ({anxiety}%)"
stress_label = f"{get_level(stress)} ({stress}%)"

if attention <= 25:
    attention_label = f"Good ({attention}%)"
elif attention <= 50:
    attention_label = f"Low ({attention}%)"
elif attention <= 75:
    attention_label = f"Very Low ({attention}%)"
else:
    attention_label = f"Extremely Low ({attention}%)"
```

```python
prompt = ChatPromptTemplate.from_template("""
You are an AI Wellness and Attention Improvement Assistant.

Assessment Results:

Depression: {depression}
Anxiety: {anxiety}
Stress: {stress}
Attention Score: {attention}
EEG Attention Score: {eeg_attention}%
Confidence Score: {confidence}%

Instructions:

1. Give a short assessment summary.
2. Explain possible factors affecting attention.
3. Recommend:
   - 3 meditation exercises
   - 3 physical activities
   - 3 focus improvement exercises
4. Create a simple daily routine.
5. Give encouraging advice.

IMPORTANT:
- Do NOT diagnose any disease.
- Do NOT prescribe medications.
- Use phrases like:
  "Your scores suggest..."
  "Your responses indicate..."
  "You may benefit from..."
- Never say:
  "You have depression"
  "You have anxiety disorder"
  "You have ADHD"

Format the response with clear headings.

""")
```

```python
chain = prompt | llm

# -----------------------------
# Invoke LLM
# -----------------------------
response = chain.invoke({
    "depression": depression_label,
    "anxiety": anxiety_label,
    "stress": stress_label,
    "attention": attention_label,
    "eeg_attention": eeg_attention,
    "confidence": confidence
})

# -----------------------------
# Print Report
# -----------------------------
print("\n")
print("=" * 80)
print("PERSONALIZED WELLNESS REPORT")
print("=" * 80)
print("\n")

print(response.content)
```

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from IPython.display import Markdown, display

# ==========================================
# Load Environment Variables
# ==========================================
load_dotenv()

# ==========================================
# Initialize Groq Model
# ==========================================
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.3,
    max_tokens=3000
)

# ==========================================
# Helper Functions
# ==========================================

def get_level(score):
    if score <= 25:
        return "Minimal / Normal"
    elif score <= 50:
        return "Moderate"
    elif score <= 75:
        return "Severe"
    else:
        return "Extremely Severe"


def get_attention_level(score):
    if score >= 75:
        return "Excellent"
    elif score >= 50:
        return "Good"
    elif score >= 25:
        return "Low"
    else:
        return "Very Low"


# ==========================================
# Sample Data
# Replace with actual questionnaire scores
# ==========================================

def get_score(metric_name):
    while True:
        try:
            score = float(input(f"Enter {metric_name} (0-100): "))

            if 0 <= score <= 100:
                return score

            print("Score must be between 0 and 100.")

        except ValueError:
            print("Please enter a valid number.")

depression = get_score("Depression Score")
anxiety = get_score("Anxiety Score")
stress = get_score("Stress Score")
attention = get_score("Attention Score")
eeg_attention = get_score("EEG Attention Score")
confidence = get_score("Confidence Score")

# ==========================================
# Labels
# ==========================================

depression_label = f"{get_level(depression)} ({depression}%)"
anxiety_label = f"{get_level(anxiety)} ({anxiety}%)"
stress_label = f"{get_level(stress)} ({stress}%)"
attention_label = f"{get_attention_level(attention)} ({attention}%)"

# ==========================================
# Primary Concern
# ==========================================

scores = {
    "Depression": depression,
    "Anxiety": anxiety,
    "Stress": stress,
    "Attention Deficit": 100 - attention
}

highest_issue = max(scores, key=scores.get)

# ==========================================
# Focus Consistency Gap
# ==========================================

focus_gap = abs(attention - eeg_attention)

# ==========================================
# Prompt
# ==========================================

prompt = ChatPromptTemplate.from_template("""
You are an AI-powered EEG Attention Assessment Coach.

Your role is to act like a personal wellness and attention improvement coach.

Assessment Results:

Depression: {depression}
Anxiety: {anxiety}
Stress: {stress}
Attention Score: {attention}
EEG Attention Score: {eeg_attention}%
Confidence Score: {confidence}%
Primary Concern: {highest_issue}
Focus Consistency Gap: {focus_gap}%

Interpretation:
- Small Gap (<10%) = questionnaire and EEG agree.
- Large Gap (>20%) = possible mismatch between perceived and measured attention.

Generate ALL sections completely.

# Assessment Summary

Create a professional table showing:

| Category | Result |
|-----------|----------|
| Depression | value |
| Anxiety | value |
| Stress | value |
| Attention | value |
| EEG Attention | value |
| Confidence | value |
| Primary Concern | value |

# Understanding Your Results

Explain:

- What each score means
- How stress may affect focus
- How anxiety may affect concentration
- What the EEG Attention Score suggests
- What the Focus Consistency Gap means

Speak directly to the user.

Example:

"Your attention score is currently lower than your EEG attention score. This may suggest that you are capable of focusing better than you currently feel."

Do NOT use clinical language.

# What's Likely Affecting Your Attention?

Explain how the following factors may influence the user's focus:

- Stress
- Anxiety
- Sleep quality
- Physical activity
- Screen time
- Study habits
- Daily routine

Keep explanations practical and easy to understand.

# Personalized Recommendations

Explain WHY each recommendation may help.

## Meditation Exercises

Provide EXACTLY 3 exercises.

For each exercise include:

- Exercise Name
- Duration
- Step-by-step Instructions
- Why it may help improve attention

## Physical Activities

Provide EXACTLY 3 activities.

For each activity include:

- Activity Name
- Duration
- Instructions
- Why it may help attention and mental performance

## Focus Improvement Activities

Provide EXACTLY 3 activities.

For each activity include:

- Activity Name
- Instructions
- Why it may help attention, concentration, or memory

# Your 7-Day Attention Improvement Plan

Create a personalized plan.

For each day include:

### Day 1
Meditation:
Physical Activity:
Focus Training:

### Day 2
Meditation:
Physical Activity:
Focus Training:

Continue until Day 7.

Make the activities slightly different each day.

# Expected Benefits

Explain:

"If you consistently follow this plan, you may notice..."

Discuss:

- Better concentration
- Improved mental clarity
- Reduced distractions
- Better productivity
- Improved confidence

# Motivational Note

Write a warm, personal message directly to the user.

Example:

"Improving attention is a skill that develops over time. Every small step you take today can contribute to better focus and mental clarity tomorrow."

# Disclaimer

Write exactly:

"This report is intended for educational and wellness purposes only and is not a medical diagnosis. If symptoms significantly affect daily life, consulting a qualified healthcare professional may be beneficial."

=================================================
IMPORTANT RULES
=================================================

1. Do NOT diagnose:
   - Depression
   - Anxiety Disorder
   - ADHD
   - Any psychiatric condition

2. Do NOT prescribe medications.

3. Do NOT skip any section.

4. Every section must be completed.

5. Focus heavily on:
   - Attention
   - Concentration
   - Cognitive performance

6. Give extra emphasis to the Primary Concern.

7. Speak directly to the user.

Use:
- "Your results suggest..."
- "You may benefit from..."
- "You might notice..."
- "Based on your responses..."
- "I recommend..."
- "To improve your attention..."

Do NOT use:
- "The individual..."
- "The participant..."
- "The assessment reveals..."
- "The user..."
- "Individuals may..."

8. Write like a supportive personal coach.

9. The report should feel like a conversation with the user, not a medical report.

10. Never refer to the user in the third person.

11. Explain WHY each recommendation is being given.

12. End the report on a positive and encouraging note.
""")

# ==========================================
# Build Chain
# ==========================================

chain = prompt | llm

# ==========================================
# Generate Report
# ==========================================

response = chain.invoke({
    "depression": depression_label,
    "anxiety": anxiety_label,
    "stress": stress_label,
    "attention": attention_label,
    "eeg_attention": eeg_attention,
    "confidence": confidence,
    "highest_issue": highest_issue,
    "focus_gap": focus_gap
})

report = response.content

# ==========================================
# Save Report
# ==========================================

with open("eeg_attention_report.txt", "w", encoding="utf-8") as file:
    file.write(report)

# ==========================================
# Display
# ==========================================

print("=" * 100)
print("EEG ATTENTION ASSESSMENT REPORT")
print("=" * 100)

print("\nTotal Characters Generated:", len(report))
print("\nReport saved to: eeg_attention_report.txt\n")

display(Markdown(report))
```

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

# ==========================================
# Load Environment Variables
# ==========================================
load_dotenv()

# ==========================================
# LLM
# ==========================================
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.3,
    max_tokens=2500
)

# ==========================================
# Helper Functions
# ==========================================

def get_score(name):
    while True:
        try:
            score = float(input(f"Enter {name} (0-100): "))
            if 0 <= score <= 100:
                return score
            print("Score must be between 0 and 100")
        except:
            print("Invalid input")

def get_level(score):
    if score <= 25:
        return "Minimal"
    elif score <= 50:
        return "Moderate"
    elif score <= 75:
        return "Severe"
    return "Extremely Severe"

# ==========================================
# Scores
# ==========================================

depression = get_score("Depression")
anxiety = get_score("Anxiety")
stress = get_score("Stress")
attention = get_score("Attention")

# ==========================================
# Primary Concern
# ==========================================

scores = {
    "Depression": depression,
    "Anxiety": anxiety,
    "Stress": stress,
    "Attention Deficit": 100 - attention
}

primary_concern = max(scores, key=scores.get)

print("\n" + "="*60)
print("ASSESSMENT SUMMARY")
print("="*60)

print(f"Depression : {get_level(depression)} ({depression}%)")
print(f"Anxiety    : {get_level(anxiety)} ({anxiety}%)")
print(f"Stress     : {get_level(stress)} ({stress}%)")
print(f"Attention  : {attention}%")
print(f"Primary Concern : {primary_concern}")

print("\nLet's understand your situation better.\n")

# ==========================================
# Dynamic Questions
# ==========================================

answers = {}

if primary_concern == "Attention Deficit":

    answers["study_hours"] = input(
        "1. How many hours do you study daily? "
    )

    answers["main_distraction"] = input(
        "2. What distracts you the most? "
        "(Mobile/Social Media/Noise/Thoughts/Other): "
    )

    answers["sleep_hours"] = input(
        "3. How many hours do you sleep daily? "
    )

    answers["screen_time"] = input(
        "4. Daily screen time in hours? "
    )

    answers["focus_method"] = input(
        "5. Do you use any focus techniques "
        "(Pomodoro/Time Blocking/None)? "
    )

elif primary_concern == "Stress":

    answers["stress_source"] = input(
        "1. Main source of stress? "
        "(Studies/Career/Family/Finance/Other): "
    )

    answers["overwhelmed"] = input(
        "2. How often do you feel overwhelmed? "
    )

    answers["sleep_hours"] = input(
        "3. How many hours do you sleep daily? "
    )

    answers["exercise"] = input(
        "4. Do you exercise regularly? "
    )

    answers["stress_management"] = input(
        "5. How do you currently manage stress? "
    )

elif primary_concern == "Anxiety":

    answers["overthinking"] = input(
        "1. Do you often overthink? "
    )

    answers["nervousness"] = input(
        "2. Do you feel nervous before exams/interviews? "
    )

    answers["racing_thoughts"] = input(
        "3. How often do you experience racing thoughts? "
    )

    answers["relaxation"] = input(
        "4. Do you practice relaxation techniques? "
    )

    answers["sleep_quality"] = input(
        "5. How would you rate your sleep quality? "
    )

else:

    answers["motivation"] = input(
        "1. Do you struggle with motivation? "
    )

    answers["social_support"] = input(
        "2. Do you have people you can talk to? "
    )

    answers["exercise"] = input(
        "3. Do you exercise regularly? "
    )

    answers["sleep_hours"] = input(
        "4. How many hours do you sleep daily? "
    )

    answers["hobbies"] = input(
        "5. What activities do you enjoy? "
    )

# ==========================================
# AI Prompt
# ==========================================

prompt = ChatPromptTemplate.from_template("""
You are an AI Attention Improvement Coach.

Assessment Results:

Depression: {depression}%
Anxiety: {anxiety}%
Stress: {stress}%
Attention: {attention}%

Primary Concern:
{primary_concern}

Follow-up Answers:
{answers}

Create a personalized report.

Speak directly to the user.

Include:

# What Your Results Suggest

# Main Factors Affecting You

# 3 Meditation Exercises

For each:
- Name
- Duration
- Instructions
- Why it helps

# 3 Physical Activities

For each:
- Duration
- Why it helps

# 3 Focus Improvement Activities

For each:
- Instructions
- Why it helps

# Personalized 7-Day Plan

Day 1 to Day 7

# Expected Benefits

# Motivational Message

Important:
- Do not diagnose diseases.
- Be supportive.
- Explain recommendations in simple language.
- Focus strongly on the primary concern.
""")

chain = prompt | llm

response = chain.invoke({
    "depression": depression,
    "anxiety": anxiety,
    "stress": stress,
    "attention": attention,
    "primary_concern": primary_concern,
    "answers": answers
})

print("\n")
print("="*70)
print("PERSONALIZED IMPROVEMENT PLAN")
print("="*70)
print("\n")

print(response.content)

with open(
    "personalized_attention_plan.txt",
    "w",
    encoding="utf-8"
) as f:
    f.write(response.content)

print("\nPlan saved to personalized_attention_plan.txt")
```
